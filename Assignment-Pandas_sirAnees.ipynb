```python
import glob
imprt_files = glob.glob('states*.csv')

dir = '../Assignment'

for path in glob.iglob(f'{dir}/states*.csv'):
    print(path)

    
```

```python
us_census = pd.concat(pd.read_csv(f) for f in imprt_files)
display(us_census.head())
display(us_census.columns)
us_census.dtypes
```

```python
us_census["Income"] = pd.to_numeric(us_census["Income"].str.replace("$","",regex=True))
```

```python
us_census[["Gender_Male","Gender_Female"]]=us_census["GenderPop"].str.split("_",expand=True)
us_census.head()
```

```python
us_census["Gender_Male"]  = pd.to_numeric(us_census["Gender_Male"].str.replace("M",""))
us_census["Gender_Female"]= pd.to_numeric(us_census["Gender_Female"].str.replace("F",""))
```

```python
us_census.drop(['GenderPop', 'Unnamed: 0'], axis = 1)
display(us_census.head())
display(us_census.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(us_census["Gender_Female"],us_census["Income"])
plt.show()
```

```python
us_census["Gender_Female"] = us_census["Gender_Female"].fillna(us_census["TotalPop"]-us_census["Gender_Male"]).astype(int)
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
plt.scatter(us_census["Gender_Female"],us_census["Income"])
plt.show()
```

```python
us_census.columns
```

```python
def replace_pc(col):
  us_census[col] = us_census[col].str.replace("%","").astype(float)
replace_pc("Hispanic")
replace_pc("White")
replace_pc("Black")
replace_pc("Native")
replace_pc("Asian")            
replace_pc("Pacific")
```
